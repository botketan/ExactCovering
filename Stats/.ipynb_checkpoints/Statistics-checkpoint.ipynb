```python
import os
import subprocess
import matplotlib.pyplot as plt
import numpy as np
import math
from matplotlib.ticker import ScalarFormatter
```

## Varying n

```python
with_vivify = []
without_vivify = []
for i in range(50, 101):
    filename = f"cadical_output{i}.tmp"
    n, k = i*10, math.floor(i*2.6)
    subprocess.run([f"echo {n} {k} | ../Problem-1/a.out" ], shell=True)
    # With the parameter
    subprocess.run([f'cadical problem_a1.txt --lucky=false | grep -A 5 "resources" | tail -n 4 | head -n 3'], shell=True, stdout=open(filename, 'w'))
    with open(filename, 'r') as f:
        lines = f.readlines()
        lines = [line.strip() for line in lines]
        lines = [line.split() for line in lines]
        lines = [item for sublist in lines for item in sublist]
        lines = [float(line) for line in lines if line.replace('.', '', 1).isdigit()]
        with_vivify.append(lines)
    # Without the parameter
    subprocess.run([f'cadical problem_a1.txt --lucky=false --vivify=false | grep -A 5 "resources" | tail -n 4 | head -n 3'], shell=True, stdout=open(filename, 'w'))
    with open(filename, 'r') as f:
        lines = f.readlines()
        lines = [line.strip() for line in lines]
        lines = [line.split() for line in lines]
        lines = [item for sublist in lines for item in sublist]
        lines = [float(line) for line in lines if line.replace('.', '', 1).isdigit()]
        without_vivify.append(lines)   

print("done")
```

```python
x_pts = [i for i in range(50 , 101)]
y1 = [h[0] for h in with_vivify]
y2 = [h[0] for h in without_vivify]
plt.plot(x_pts, y1, label="with vivify")
plt.plot(x_pts, y2, label="without vivify")

plt.yscale('log')

formatter = ScalarFormatter(useOffset=True)
formatter.set_scientific(True)
plt.gca().yaxis.set_major_formatter(formatter)

plt.xlabel('n')
plt.title('n vs time')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)  # Add grid lines
plt.show()
```

## Varying k

```python
with_vivify_k = []
without_vivify_k = []
for i in range(1, 60):
    filename = f"cadical_output{i}.tmp"
    n, k = 80, i
    subprocess.run([f"echo {n} {k} | ../Problem-1/a.out" ], shell=True)
    #With the paramter
    subprocess.run([f'cadical problem_a1.txt --lucky=false --vivifymaxeff=2e9 --vivifymineff=0 | grep -A 5 "resources" | tail -n 4 | head -n 3'], shell=True, stdout=open(filename, 'w'))
    with open(filename, 'r') as f:
        lines = f.readlines()
        lines = [line.strip() for line in lines]
        lines = [line.split() for line in lines]
        lines = [item for sublist in lines for item in sublist]
        lines = [float(line) for line in lines if line.replace('.', '', 1).isdigit()]
        with_vivify_k.append(lines)
    #Without the parameter
    subprocess.run([f'cadical problem_a1.txt --lucky=false  | grep -A 5 "resources" | tail -n 4 | head -n 3'], shell=True, stdout=open(filename, 'w'))
    with open(filename, 'r') as f:
        lines = f.readlines()
        lines = [line.strip() for line in lines]
        lines = [line.split() for line in lines]
        lines = [item for sublist in lines for item in sublist]
        lines = [float(line) for line in lines if line.replace('.', '', 1).isdigit()]
        without_vivify_k.append(lines)   

print("done")
```

```python
x_pts_k = [i for i in range(1,60)]
y1_k = [h[0] for h in with_vivify_k]
y2_k = [h[0] for h in without_vivify_k]
plt.plot(x_pts_k, y1_k, label="with vivify")
plt.plot(x_pts_k, y2_k, label="without vivify")

plt.yscale('log')

formatter = ScalarFormatter(useOffset=True)
formatter.set_scientific(True)
plt.gca().yaxis.set_major_formatter(formatter)

plt.xlabel('n')
plt.title('n vs time')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)  # Add grid lines
plt.show()
```
