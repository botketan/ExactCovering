# Effect of `vivify`, `arenatype` and `cover`

```python
import subprocess
import re
import pandas as pd
import matplotlib.pyplot as plt
from concurrent.futures import ThreadPoolExecutor, as_completed
import os
from tqdm import tqdm
```

```python
# Define the x-axis parameter and its values
x_axis_name = 'arenatype'
x_axis_param = {x_axis_name: ['1', '2', '3']}

parameter_grid = {
    'binary': ['true', 'false'],
    'block': ['true', 'false'],
    'check': ['true', 'false'],
    'checkassumptions': ['true', 'false'],
    'checkconstraint': ['true', 'false'],
    'checkfrozen': ['true', 'false'],
    'checkfailed': ['true', 'false'],
    'checkproof': ['true', 'false'],
    'checkprooflrat': ['true', 'false'],
    'checkwitness': ['true', 'false'],
    'chrono': ['1','2','3'],
    'chronoalways': ['true', 'false'],
    'chronoreusetrail': ['true', 'false'],
    'compact': ['true', 'false'],
    'condition': ['true', 'false'],
    'cover': ['true', 'false'],
    'decompose': ['true', 'false'],
    'deduplicate': ['true', 'false'],
    'eagersubsume': ['true', 'false'],
    'elim': ['true', 'false'],
    'elimands': ['true', 'false'],
    'elimbackward': ['true', 'false'],
    'elimequivs': ['true', 'false'],
    'elimites': ['true', 'false'],
    'elimlimited': ['true', 'false'],
    'elimsubst': ['true', 'false'],
    'elimxors': ['true', 'false'],
    'flush': ['true', 'false'],
    'forcephase': ['true', 'false'],
    'inprocessing': ['true', 'false'],
    'instantiate': ['true', 'false'],
    'instantiateonce': ['true', 'false'],
    'lratexternal': ['true', 'false'],
    'lratfrat': ['true', 'false'],
    'lratveripb': ['true', 'false'],
    'lucky': ['true', 'false'],
    'minimize': ['true', 'false'],
    'otfs': ['true', 'false'],
    'phase': ['true', 'false'],
    'probe': ['true', 'false'],
    'probehbr': ['true', 'false'],
    'realtime': ['true', 'false'],
    'reduce': ['true', 'false'],
    'restoreall': ['0', '1', '2'],
    'restoreflush': ['true', 'false'],
    'reverse': ['true', 'false'],
    'score': ['true', 'false'],
    'shuffle': ['true', 'false'],
    'shufflequeue': ['true', 'false'],
    'shufflerandom': ['true', 'false'],
    'shufflescores': ['true', 'false'],
    'stabilize': ['true', 'false'],
    'stabilizeonly': ['true', 'false'],
    'subsume': ['true', 'false'],
    'subsumelimited': ['true', 'false'],
    'subsumestr': ['true', 'false'],
    'ternary': ['true', 'false'],
    'transred': ['true', 'false'],
    'verbose': ['0', '1', '2', '3'],
    'vivify': ['true', 'false'],
    'vivifyinst': ['true', 'false'],
    'vivifyonce': ['0', '1', '2'],
    'walk': ['true', 'false'],
    'walknonstable': ['true', 'false'],
    'walkredundant': ['true', 'false']
}


metrics_patterns = {
    'restarts': r"c restarts:\s+(\d+)",
    'conflicts': r"c conflicts:\s+(\d+)",
    'time': r"c total process time since initialization:\s+(\d+\.\d+)"
}

max_workers = os.cpu_count()
```

```python
#This will take a long time

def run_cadical(command, metrics_patterns, callback, arena_type, param_name, param_value):
  """Runs the cadical command and calls the callback with extracted data."""
  result = subprocess.run(command, shell=True, capture_output=True, text=True)
  output = result.stdout
  result_data = {x_axis_name: arena_type, param_name: param_value }

  for metric, pattern in metrics_patterns.items():
    match = re.search(pattern, output)
    if match:
      result_data[metric] = float(match.group(1)) if '.' in match.group(1) else int(match.group(1))
    else:
      result_data[metric] = None

  callback(result_data)

with ThreadPoolExecutor(max_workers=max_workers) as executor: 
  futures = []
  results = []

  for fixed_arenatype in x_axis_param[x_axis_name]:
    for param_name, param_values in parameter_grid.items():
      for param_value in param_values:
        command = f"cadical problem_a1.txt --lucky=false --{x_axis_name}={fixed_arenatype} --{param_name}={param_value}"
        #print(f"Running: {command}")
        future = executor.submit(run_cadical, command, metrics_patterns, results.append, fixed_arenatype, param_name, param_value)
        futures.append(future)

  for future in tqdm(as_completed(futures),total=len(futures)):
    future.result() 

  df = pd.DataFrame(results)
```

```python
def plot_metrics(df, x_axis_param_name, metric_to_plot, grouping_param):
    plt.figure(figsize=(5, 3))

    for group_value in df[grouping_param].unique():
        subset = df[df[grouping_param] == group_value]
        
        plt.plot(subset[x_axis_param_name], subset[metric_to_plot], 
                 marker='o', linestyle='-', 
                 label=f'{grouping_param}: {group_value}')

    plt.title(f"{metric_to_plot.capitalize()} vs {x_axis_param_name.capitalize()} by {grouping_param.capitalize()} Value")
    plt.xlabel(x_axis_param_name.capitalize())
    plt.ylabel(metric_to_plot.capitalize())
    plt.xticks(subset[x_axis_param_name])
    plt.legend(title=grouping_param.capitalize())
    plt.grid()
    plt.tight_layout()
    plt.show()
```

```python
for key in parameter_grid:
    plot_metrics(df, x_axis_name, 'time', key)
```

```python
def compute_sat(param, values):
  """Computes SAT and returns a DataFrame using futures."""
  results = []
  with ThreadPoolExecutor(max_workers=max_workers) as executor:
    futures = []

    for value in values:
      command = f"cadical problem_a1.txt --lucky=false --{param}={value} --arenatype=3 --vivify=false --eagersubsume=true --subsumestr=false --stabilize=false --phase=false"
      future = executor.submit(run_cadical, command, metrics_patterns, results.append, 3, param, value)
      futures.append(future)

    for future in tqdm(as_completed(futures),total=len(futures),desc=param):
      future.result()

  return pd.DataFrame(results)
```

```python
def plot_sat(df, param, metric_to_plot):
    plt.figure(figsize=(10, 6))
    plt.plot(df[metric_to_plot], df[param], marker='o', linestyle='-')
    plt.title(f"{param.capitalize()} vs {metric_to_plot.capitalize()}")
    plt.xlabel(metric_to_plot.capitalize())
    plt.ylabel(param.capitalize())
    plt.grid()
    plt.tight_layout()
    plt.show()
```

```python
fixed_params = {
    'arenatype': 3,
    'vivify': 'false',
    'eagersubsume': 'true',
    'subsumestr': 'false',
    'stabilize': 'false',
    'phase': 'false',
}

df2 = pd.DataFrame()
params = {
    'flushfactor': range(1,1001,9),
    #'scorefactor': range(500,1001,10)
}

for param, values in params.items():
    df_param = compute_sat(param, values, fixed_params)
    df2 = pd.concat([df2, df_param])
```
